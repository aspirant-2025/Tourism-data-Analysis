```python
grouping = df.groupby('Country')['Visitors'].max()
```

```python
grouping
```

```python
top_5 = grouping.sort_values(ascending = False).nlargest(5)
```

```python
top_5
```

```python
#converting into dictionary
result = [{'Country': country, 'Total_Visitors': visitors} 
          for country, visitors in top_5.items()]
```

```python
result
```

Average rating of each category

```python
avg_rating  = df.groupby('Category')['Rating'].mean()     
```

```python
avg_rating
```

```python
result  = [{'Category Names':Category , 'Average Rating':round(Rating , 2)} for Category, Rating in avg_rating.items()]
```

```python
result
```
